```python
new_df = pd.DataFrame();
gws = np.arange(1, 39)
for gw in gws:
    gw_df = pd.read_csv("gw{}.csv".format(gw), encoding = "ISO-8859-1")
    gw_df["GW"] = gw
    new_df = new_df.append(gw_df)
```

```python
new_df.reset_index(drop=True, inplace=True)
```

```python
new_df
```

```python
def getResult(row):
    if (row["team_h_score"] == row["team_a_score"]):
        return "D";
    elif (row["was_home"] is True and row["team_h_score"] < row["team_a_score"]):
        return "L";
    elif (row["was_home"] is True and row["team_h_score"] > row["team_a_score"]):
        return "W";
    elif (row["was_home"] is not True and row["team_h_score"] < row["team_a_score"]):
        return "W";
    elif (row["was_home"] is not True and row["team_h_score"] > row["team_a_score"]):
        return "L";
```

```python
df["result_pgw"] = new_df.apply(lambda row: getResult(row), axis=1)
```

```python
df.drop(["opponent_team"], inplace=True, axis=1)
```
