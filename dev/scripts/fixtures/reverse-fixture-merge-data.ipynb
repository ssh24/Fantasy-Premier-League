```python
to_remove_cols = ["minutes", "goals_scored", "goals_conceded", "assists", "clean_sheets", "own_goals", 
                  "yellow_cards", "red_cards", "penalties_saved", "penalties_missed", "saves", "bonus", "bps",
                 "creativity", "influence", "threat", "ict_index"];
```

```python
non_remove_cols = ["result", "total_points"]
```

```python
all_cols = to_remove_cols + non_remove_cols;
```

```python
player_ids = np.unique(df["player_id"]);
```

```python
opponent_ids = np.unique(df["opponent_id"]);
```

```python
at_homes = np.unique(df["at_home"]);
```

```python
for col in to_remove_cols:
    df[col+"_rlf"] = None
    df[col+"_rsf"] = None
```

```python
def augmentReverseFixtureData(df, rf_type, player, opponent, at_home=1):
    if (rf_type == "rlf"):
        sub_df = df[(df["player_id"] == player) & (df["opponent_id"] == opponent)]
    elif (rf_type == "rsf"):
        sub_df = df[(df["player_id"] == player) & (df["opponent_id"] == opponent) & 
                    (df["at_home"] == at_home)]
    else:
        return "Undefined reverse fixture type";
    
    indexes = list(sub_df.index)
    for i in range(len(indexes)):
        index = indexes[i];
        for target in all_cols:
            if (i == 0):
                df.ix[index, "{}_{}".format(target, rf_type)] = 0
            else:
                df.ix[index, "{}_{}".format(target, rf_type)] = df.iloc[indexes[i-1]][target];
    
    return df;
```

```python
for player in player_ids:
    for opponent in opponent_ids:
        # work on RLF data
        df = augmentReverseFixtureData(df, "rlf", player, opponent);
        
        for at_home in at_homes:
            # work on RSF data
            df = augmentReverseFixtureData(df, "rsf", player, opponent, at_home);
    
    print("Done with player {}".format(player));
```
