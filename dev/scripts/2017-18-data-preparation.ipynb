```python
import os
import pandas as pd
import numpy as np
import time
```

```python
target_season = "2017-18"
```

```python
os.chdir(os.path.join(os.getcwd(), "..", "..", "data", target_season, "gws"));
```

```python
teams = ["Arsenal", "Liverpool", "Chelsea", "Bournemouth", "Crystal Palace", "Manchester City", "Watford",
        "Manchester United", "Tottenham Hotspur", "Everton", "West Bromwich Albion", "Southampton",
        "Leicester City", "Newcastle United", "Brighton and Hove Albion", "Swansea City", "Stoke City",
        "Huddersfield Town", "West Ham United", "Burnley"]
```

```python
teams.sort()
```

```python
champion_ls = "Chelsea"
```

```python
top5_ls = ["Chelsea", "Tottenham Hotspur", "Manchester City", "Liverpool", "Arsenal"]
```

```python
bottom5_ls = ["Stoke City", "Crystal Palace", "Swansea City", "Burnley", "Watford"]
```

```python
promoted_ts = ["Newcastle United", "Brighton and Hove Albion", "Huddersfield Town"];
```

```python
df = pd.read_csv("merged_gw.csv");
```

```python
df["season"] = 2017
```

```python
df.columns
```

```python
os.chdir(os.path.join(os.getcwd(), "..", "..", "..", "dev", "data"));
```

```python
team_id_df = pd.read_csv("team_id_map.csv");
```

```python
player_id_df = pd.read_csv("player_id_map.csv");
```

```python
# set team from the list of teams that season
df["team"] = df["opponent_team"].apply(lambda team: teams[team-1])
```

```python
# encode the teams into id
df["opponent_id"] = df["team"].apply(lambda team: team_id_df[team_id_df["team_name"] == team]["team_id"].iloc[0])
```

```python
# set column for is champions last season
df["champion_ls"] = df["team"].apply(lambda team: 1 if team == champion_ls else 0)
```

```python
# set column for top 5 last season
df["top5_ls"] = df["team"].apply(lambda team: 1 if team in top5_ls else 0)
```

```python
# set column for bottom 5 last season
df["bottom5_ls"] = df["team"].apply(lambda team: 1 if team in bottom5_ls else 0)
```

```python
# set column for promoted teams this season
df["promoted_ts"] = df["team"].apply(lambda team: 1 if team in promoted_ts else 0)
```

```python
# drop the opponent team and team columns
df.drop(["team", "opponent_team"], axis=1, inplace=True)
```

```python
# encode the players into id
df["player_id"] = df["name"].apply(lambda player: player_id_df[player_id_df["player_name"] == player]["player_id"].iloc[0])
```

```python
# encode was_home into labels
df["at_home"] = df["was_home"].apply(lambda at_home: 0 if not at_home else 1)
```

```python
# drop the player name and was home columns
df.drop(["name", "was_home"], axis=1, inplace=True)
```

```python
# drop the id column
df.drop(["id"], axis=1, inplace=True)
```

```python
target_col = ["total_points"]
```

```python
cols = ["season", "GW", "at_home", "player_id", "opponent_id", "champion_ls", "top5_ls", "bottom5_ls", "promoted_ts",
        "influence", "creativity", "threat", "ict_index"]
```

```python
to_remove_cols = ["minutes", "goals_scored", "goals_conceded", "assists", "clean_sheets", "own_goals", "yellow_cards", "red_cards", "penalties_conceded", "penalties_saved", "penalties_missed",
       "saves", "dribbles", "attempted_passes", "bonus", "bps", "big_chances_created", "big_chances_missed",
       "clearances_blocks_interceptions", "completed_passes", "errors_leading_to_goal", "errors_leading_to_goal_attempt",
       "fouls", "key_passes", "offside", "open_play_crosses", "recoveries", "target_missed"]
```

```python
cols = cols + to_remove_cols + target_col;
```

```python
season = df[cols]
```

```python
season.sort_values(["GW"], inplace=True)
```

```python
season.reset_index(inplace=True, drop=True)
```

```python
season
```

```python
season.columns
```

```python
target_pgw = ["total_points", "ict_index", "influence", "creativity", "threat"];
target_pgw = target_pgw + to_remove_cols;
player_ids = np.unique(season["player_id"]);
new_df = pd.DataFrame(columns=list(season.columns));
```

```python
for player_id in player_ids:
    now = time.time();
    df_ = season[season["player_id"] == player_id]
    df_.reset_index(inplace=True, drop=True)
    for index in range(len(df_)):
        instance = df_.iloc[index]
        gw = instance["GW"];
        for target in target_pgw:
            if (gw == 1 or index == 0):
                df_.ix[index, "{}_pgw".format(target)] = 0
            else:
                df_.ix[index, "{}_pgw".format(target)] = df_.iloc[index-1][target]
    new_df = new_df.append(df_)
    later = time.time();
    difference = int(later - now);
    print("Done with player {} --- Time taken: {}s".format(player_id, difference))
```

```python
# -1 means L
# 0 means D
# 1 means W
def getResult(row):
    if (row["team_h_score"] == row["team_a_score"]):
        return 0;
    elif (row["at_home"] == 1 and row["team_h_score"] < row["team_a_score"]):
        return -1;
    elif (row["at_home"] == 1 and row["team_h_score"] > row["team_a_score"]):
        return 1;
    elif (row["at_home"] == 0 and row["team_h_score"] < row["team_a_score"]):
        return 1;
    elif (row["at_home"] == 0 and row["team_h_score"] > row["team_a_score"]):
        return -1;
```

```python
new_df["result_pgw"] = df.apply(lambda row: getResult(row), axis=1)
```

```python
new_df.reset_index(inplace=True, drop=True)
```
