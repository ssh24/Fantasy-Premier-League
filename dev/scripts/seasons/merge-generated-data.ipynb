```python
df16 = pd.read_csv("2016-prepared-data.csv");
df17 = pd.read_csv("2017-prepared-data.csv");
df18 = pd.read_csv("2018-prepared-data.csv");
df19 = pd.read_csv("2019-prepared-data.csv");
```

```python
df_merge = pd.concat([df16, df17, df18, df19])
```

```python
df_merge.sort_values(["season", "GW"], inplace=True);
```

```python
## only for duplicated player ids. this should be removed if the data preparation script is run with old season
df_merge.loc[(df_merge["player_id"] == 744), 'player_id'] = 201;
df_merge.loc[(df_merge["player_id"] == 948), 'player_id'] = 176;
```
