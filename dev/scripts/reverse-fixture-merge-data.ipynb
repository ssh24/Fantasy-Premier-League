```python
import os
import pandas as pd
import numpy as np
import warnings
import time
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
os.chdir(os.path.join(os.getcwd(), "..", "data"));
```

```python
df2016 = pd.read_csv("2016-17-prepared-data.csv");
df2017 = pd.read_csv("2017-18-prepared-data.csv");
```

```python
df_merge = pd.concat([df2016, df2017])
```

```python
to_remove_cols = ["minutes", "goals_scored", "goals_conceded", "assists", "clean_sheets", "own_goals", "yellow_cards", "red_cards", "penalties_conceded", "penalties_saved", "penalties_missed",
       "saves", "dribbles", "attempted_passes", "bonus", "bps", "big_chances_created", "big_chances_missed",
       "clearances_blocks_interceptions", "completed_passes", "errors_leading_to_goal", "errors_leading_to_goal_attempt",
       "fouls", "key_passes", "offside", "open_play_crosses", "recoveries", "target_missed", "total_points"]
```

```python
player_ids = np.unique(df_merge["player_id"]);
```

```python
opponent_ids = np.unique(df_merge["opponent_id"]);
```

```python
at_homes = np.unique(df_merge["at_home"]);
```

```python
new_df = pd.DataFrame(columns=list(df_merge.columns));
rlf_df = pd.DataFrame(columns=list(df_merge.columns));
rsf_df = pd.DataFrame(columns=list(df_merge.columns));
```

```python
for player in player_ids:
    now = time.time();
    for opponent in opponent_ids:
        df_rlf = df_merge.loc[(df_merge["player_id"] == player) & (df_merge["opponent_id"] == opponent)]
        df_rlf.reset_index(inplace=True, drop=True)
        for index in range(len(df_rlf)):
            instance = df_rlf.iloc[index];
            for target in to_remove_cols:
                if (index == 0):
                    df_rlf.ix[index, "{}_rlf".format(target)] = 0
                else:
                    df_rlf.ix[index, "{}_rlf".format(target)] = df_rlf.iloc[index-1][target]
        rlf_df = rlf_df.append(df_rlf);
    later = time.time();
    difference = int(later - now);
    print("Done with player {} --- Time taken: {}s".format(player, difference))
```

```python
for player in player_ids:
    now = time.time();
    for opponent in opponent_ids:
        for at_home in at_homes: 
            df_ = df_merge.loc[(df_merge["player_id"] == player) & (df_merge["opponent_id"] == opponent) & (df_merge["at_home"] == at_home)]
            df_.reset_index(inplace=True, drop=True)
            for index in range(len(df_)):
                instance = df_.iloc[index];
                for target in to_remove_cols:
                    if (index == 0):
                        df_.ix[index, "{}_rsf".format(target)] = 0
                    else:
                        df_.ix[index, "{}_rsf".format(target)] = df_.iloc[index-1][target]
            rsf_df = rsf_df.append(df_);
    later = time.time();
    difference = int(later - now);
    print("Done with player {} --- Time taken: {}s".format(player, difference))
```

```python
rsf_df.reset_index(inplace=True, drop=True);
rlf_df.reset_index(inplace=True, drop=True);
```
