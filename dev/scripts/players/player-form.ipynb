```python
import os
import pandas as pd
import numpy as np
import warnings
import time
from scipy import stats
```

```python
os.chdir(os.path.join(os.getcwd(), "..", "..", "data"));
```

```python
df = pd.read_csv("combined-season-data.csv");
```

```python
df.columns
```

```python
df.head()
```

```python
num_gws = 3;
```

```python
player_ids = np.unique(df["player_id"])
```

```python
for player in player_ids:
    sub_df = df[df["player_id"] == player];
    indexes = list(sub_df.index)
    
    for i in range(len(indexes)):
        index = indexes[i]
        
        if (i < num_gws):
            df.ix[index, "player_form"] = 0
        else:
            avg_points_player = np.mean(sub_df[(i-num_gws):i]["total_points"])
            points_overall = [];
            
            for ind, row in sub_df[(i-num_gws):i].iterrows():
                season = row["season"];
                gw = row["GW"];
                avg_point_ov = np.mean(df[(df["season"] == season) & (df["GW"] == gw)]["total_points"])
                points_overall.append(avg_point_ov);
            
            avg_points_overall = np.mean(points_overall);
            df.ix[index, "player_form"] = (avg_points_player / avg_points_overall);
            
    print("Done player {}".format(player))
```

```python
df_cols = list(np.unique(df.columns))
```

```python
main_cols = ['season', 'GW', 'at_home', 'player_id', 'team_id', 'opponent_id', 'champion_ls',
       'top5_ls', 'bottom5_ls', 'promoted_ts', 'total_points']
```

```python
target_col = ["total_points"]
```

```python
cols = [item for item in df_cols if item not in main_cols]
```

```python
final_cols = main_cols + cols + target_col
```
