```python
df16 = pd.read_csv("2016-prepared-data.csv");
df17 = pd.read_csv("2017-prepared-data.csv");
df18 = pd.read_csv("2018-prepared-data.csv");
df19 = pd.read_csv("2019-prepared-data.csv");
```

```python
df_merge = pd.concat([df16, df17, df18, df19])
```

```python
to_remove_cols = ["minutes", "goals_scored", "goals_conceded", "assists", "clean_sheets", "own_goals", 
                  "yellow_cards", "red_cards", "penalties_saved", "penalties_missed", "saves", "bonus", "bps",
                 "creativity", "influence", "threat", "ict_index", "result"];
```

```python
df_merge.drop(to_remove_cols, axis=1, inplace=True)
```

```python
df_merge.sort_values(["season", "GW"], inplace=True);
```
