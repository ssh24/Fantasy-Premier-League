```python
import os
import pandas as pd
import numpy as np
```

```python
os.chdir(os.path.join(os.getcwd(), "..", "data"));
```

```python
target_gw = 7;
```

```python
target_season = "2019";
```

```python
df = pd.read_csv("{}-prepared-data.csv".format(target_season))
```

```python
df.sort_values(["season", "GW"], inplace=True)
```

```python
df.tail()
```

```python
player_id_df = pd.read_csv("player_id_map.csv");
```

```python
team_id_df = pd.read_csv("team_id_map.csv");
```

```python
players = list(player_id_df["player_id"]);
```

```python
new_df = pd.DataFrame(columns=df.columns)
```

```python
for player in players:
    df_ = df[(df["player_id"] == player) & (df["GW"] == target_gw-1)];
    new_df = new_df.append(df_)
```

```python
main_cols = ['at_home', 'player_id', 'opponent_id', 'champion_ls',
       'top5_ls', 'bottom5_ls', 'promoted_ts'];
```

```python
cols = ['assists', 'bonus',
       'bps', 'clean_sheets',
        'creativity',
       'goals_conceded', 'goals_scored', 
       'ict_index', 'influence', 'minutes',
       'own_goals', 'penalties_missed',
       'penalties_saved', 'red_cards',
       'saves', 
       'threat','yellow_cards', "result", 'total_points']
```

```python
cols_to_keep = main_cols + cols
```

```python
new_df = new_df[cols_to_keep]
```

```python
for col in cols:
    print("Performing act on column: {}".format(col))
    new_df.rename(columns={col: col+"_pgw"}, inplace=True)
```

```python
new_df.reset_index(inplace=True, drop=True)
```

```python
new_df
```

```python
new_df.to_csv("all-players-gw{}-fpl-data.csv".format(target_gw), index=False)
```
