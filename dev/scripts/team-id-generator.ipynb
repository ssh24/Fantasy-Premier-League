```python
team_names = [];
team_ids = [];
```

```python
for season in teams_in_season:
    teams = teams_in_season[season]
    teams.sort()
    for team in teams:
        if not (team in team_names):
            team_names.append(team);
            team_ids.append(len(team_ids) + 1);
```

```python
df = pd.DataFrame({"team_name": team_names, "team_id": team_ids });
```
